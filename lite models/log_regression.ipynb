```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

np.random.seed(59)

# set the max columns to none
pd.set_option('display.max_columns', None)

%load_ext pycodestyle_magic
%pycodestyle_on
```

```python
full_path = 'D:\\machine learning\\ml projects\\'
prj = 'Main Pet Project for May-June 2022\\Customer Churn Prediction\\'
rel_path = 'featured_data\\train.csv'
df = pd.read_csv(full_path + prj + rel_path)
df.head()
```

```python
X = df.drop('churn', axis=1)
y = df.churn
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
# prepare the cross-validation procedure
cv = StratifiedKFold(n_splits=10, random_state=59, shuffle=True)
# create model
model = LogisticRegression()
# evaluate model
scores = cross_val_score(model, X, y, scoring='f1_weighted', cv=cv, n_jobs=-1)
# report performance
print('Weighted F1-Score: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
# Grid search cross validation
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

grid = {"C": np.logspace(-3, 3, 7),
        "penalty": ["l1", "l2"]}    # l1 lasso l2 ridge

logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg, grid, cv=cv,
                         scoring='f1_weighted')
logreg_cv.fit(X, y)

print("Best params:", logreg_cv.best_params_)
print('Weighted F1-score:', logreg_cv.best_score_)
```
