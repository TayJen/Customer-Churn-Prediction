```python
import autosklearn
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

np.random.seed(59)

# set the max columns to none
pd.set_option('display.max_columns', None)
```

```python
full_path = '/content/drive/MyDrive/Churn Prediction - Kaggle/Featured data/train.csv'
df = pd.read_csv(full_path)
df.head()
```

```python
X = df.drop('churn', axis=1)
y = df.churn
```

```python
from collections import Counter
from imblearn.combine import SMOTEENN

# summarize class distribution
counter = Counter(y)
print(counter)
# transform the dataset
oversample = SMOTEENN()
X, y = oversample.fit_resample(X, y)
# summarize the new class distribution
counter = Counter(y)
print(counter)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=59)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import autosklearn.classification

autosklearn_classifier = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=240,
    per_run_time_limit=60
)
```

```python
autosklearn_classifier.fit(X_train, y_train)
```

```python
autosklearn_classifier.leaderboard()
```

```python
from sklearn.metrics import classification_report, f1_score


y_pred = autosklearn_classifier.predict(X_test)
print(classification_report(y_test, y_pred))
print(f1_score(y_test, y_pred))
```
